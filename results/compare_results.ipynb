```python
import pandas as pd
import matplotlib.pyplot as plt
import os

def list_files(ext=".csv"):
    files = os.listdir('.')
    print("Files in the current directory:")
    for file in files:
        if file.endswith(ext):
            print(file)

def load_dfs(files):
    dfs = {}
    for file in files:
        if file.endswith('.csv'):
            df = pd.read_csv(file, index_col=0)
            dfs["".join("".join(file.split("_")[-2:]).split(".")[0:-1])] = df

    dfs = pd.concat(dfs, axis=1, keys=dfs.keys())
    return dfs

dfs = load_dfs(os.listdir('.'))
```

```python
dfs
```

```python
filter_columns = [col for col in dfs.columns if "1" in col[0]]
filter_columns = [col for col in filter_columns if "mu" in col[0] or "NTK" in col[0]]
dfs = dfs[filter_columns]
dfs
```

```python
import matplotlib.colors as mcolors

# Plot style setup
plt.style.use("seaborn-v0_8-whitegrid")  # clean background

fig, ax = plt.subplots(figsize=(10, 6))

color_map = {
    'muMLP10': (0.8, 0.1, 0.1),  
    'NTKMLP10': (0.1, 0.1, 0.8),  
}

sub_keys = sorted(set(col[1] for col in dfs.columns))
subkey_ints = {k: int(''.join(filter(str.isdigit, k))) for k in sub_keys}
norm = mcolors.Normalize(vmin=min(subkey_ints.values()), vmax=max(subkey_ints.values()))

for col in dfs.columns:
    model_type, size_str = col
    base_color = color_map[model_type]

    width = subkey_ints[size_str]
    hsv = mcolors.rgb_to_hsv(base_color)
    hsv[2] = norm(width) * 0.5 + 0.5 
    color = mcolors.hsv_to_rgb(hsv)

    model_name = model_type.replace("MLP10", "")
    label = f"{model_name} {size_str}"
    dfs[col].plot(ax=ax, label=label, color=color, linewidth=2)

ax.set_title("muP vs SP\nTraining Loss vs Learning Rate (CIFAR-10, MLP)", fontsize=16, weight='bold')
ax.set_xlabel("log2(Learning Rate)", fontsize=14)
ax.set_ylabel("Training Loss (log scale)", fontsize=14)
ax.set_yscale('log')

ax.tick_params(axis='both', which='major', labelsize=12)
ax.grid(True, which='both', linestyle='--', linewidth=0.5)

ax.legend(title="Model Variant", fontsize=11, title_fontsize=12, loc='lower right', frameon=False)

plt.tight_layout()
plt.savefig("mlp_cifar10_ready.png", dpi=300)
plt.show()
```
